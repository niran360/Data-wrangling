```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
cell_url = "https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/cell_phones.csv"
pop_url = "https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/population1.csv"
geo_url = "https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--entities--geo--country.csv"
```

```python
df_cell = pd.read_csv(cell_url)
df_pop = pd.read_csv(pop_url)
df_geo = pd.read_csv(geo_url)
```

```python
df_cell_pop = pd.merge(df_cell,df_pop,on=['Country',"year"])
```
